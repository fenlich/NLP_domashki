```python
import spacy
nlp = spacy.load("ru_core_news_sm")
doc = nlp(text)
```

```python
spacy_lemmas = []
for sent in doc.sents: # достаем предложения
    for token in sent: # достаем токены
        spacy_lemmas.append(token.lemma_)
print(spacy_lemmas)        
```

## Различия
Как минимум, spacy не токенизирует пробелы, но токенизирует некоторые пробельные символы (\n). Лемматизирует глаголы в инфинитив совершенного вида.
